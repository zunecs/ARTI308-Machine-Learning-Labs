# Matplotlib Object Oriented Method
Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.

## Introduction to the Object Oriented Method

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create Figure (empty canvas)
fig = plt.figure() # Create a new, empty figure object

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Add axes to the figure, specifying [left, bottom, width, height] as fractions of the figure width and height

# Plot on that set of axes
axes.plot(x, y, 'b') # Plot x vs y with a blue line ('b') on the created axes
axes.set_xlabel('Set X Label') # Set the label for the X-axis of these axes
axes.set_ylabel('Set y Label') # Set the label for the Y-axis of these axes
axes.set_title('Set Title') # Set the title for these axes
```

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:

```python
# Creates blank canvas
fig = plt.figure() # Create a new, empty figure object

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Add the main set of axes to the figure
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Add an inset set of axes to the figure (left, bottom, width, height)

# Larger Figure Axes 1
axes1.plot(x, y, 'b') # Plot x vs y with a blue line on the first set of axes
axes1.set_xlabel('X_label_axes1') # Set the X-axis label for axes1
axes1.set_ylabel('Y_label_axes1') # Set the Y-axis label for axes1
axes1.set_title('Axes 1 Title') # Set the title for axes1

# Insert Figure Axes 2
axes2.plot(y, x, 'r') # Plot y vs x with a red line on the second set of axes
axes2.set_xlabel('X_label_axes2') # Set the X-axis label for axes2
axes2.set_ylabel('Y_label_axes2') # Set the Y-axis label for axes2
axes2.set_title('Axes 2 Title'); # Set the title for axes2
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots() # Create a figure and a single set of axes, unpacking them into fig and axes

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r') # Plot x vs y with a red line on the created axes
axes.set_xlabel('x') # Set the X-axis label
axes.set_ylabel('y') # Set the Y-axis label
axes.set_title('title'); # Set the title of the plot
```

Then you can specify the number of rows and columns when creating the subplots() object:

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2) # Create a figure and a 1x2 grid of subplots, unpacking them into fig and an array of axes
```

```python
# Axes is an array of axes to plot on
axes # Display the array of axes objects
```

We can iterate through this array:

```python
for ax in axes: # Iterate through each axes object in the 'axes' array
    ax.plot(x, y, 'b') # Plot x vs y with a blue line on the current axes
    ax.set_xlabel('x') # Set the X-axis label for the current axes
    ax.set_ylabel('y') # Set the Y-axis label for the current axes
    ax.set_title('title') # Set the title for the current axes

# Display the figure object
fig # Display the figure containing the subplots
```

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # Create a figure and a 1x2 grid of subplots

for ax in axes: # Iterate through each axes object in the 'axes' array
    ax.plot(x, y, 'g') # Plot x vs y with a green line on the current axes
    ax.set_xlabel('x') # Set the X-axis label for the current axes
    ax.set_ylabel('y') # Set the Y-axis label for the current axes
    ax.set_title('title') # Set the title for the current axes

fig # Display the figure (before tight_layout applies)
plt.tight_layout() # Automatically adjust subplot parameters for a tight layout, preventing labels from overlapping
```

### Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments.
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch).

For example:

```python
fig = plt.figure(figsize=(8,4), dpi=100) # Create a new figure with a specified size (width=8, height=4 inches) and a resolution of 100 dots per inch
```

The same arguments can also be passed to layout managers, such as the `subplots` function:

```python
fig, axes = plt.subplots(figsize=(4,2)) # Create a figure and a single set of axes, with a specified figure size

axes.plot(x, y, 'r') # Plot x vs y with a red line on the axes
axes.set_xlabel('x') # Set the X-axis label
axes.set_ylabel('y') # Set the Y-axis label
axes.set_title('title'); # Set the title of the plot
```

## Legends, labels and titles

Now that we have covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how decorate a figure with titles, axis labels, and legends.

**Figure titles**

A title can be added to each axis instance in a figure. To set the title, use the `set_title` method in the axes instance:

```python
ax.set_title("title"); # Set the title of the current axes object
```

**Axis labels**

Similarly, with the methods `set_xlabel` and `set_ylabel`, we can set the labels of the X and Y axes:

```python
ax.set_xlabel("x") # Set the X-axis label of the current axes object
ax.set_ylabel("y"); # Set the Y-axis label of the current axes object
```

### Legends

You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure:

```python
fig = plt.figure() # Create a new, empty figure object

ax = fig.add_axes([0,0,1,1]) # Add axes to the figure, covering the entire figure area [left, bottom, width, height]

ax.plot(x, x**2, label="x**2") # Plot x vs x squared, assigning a label for the legend
ax.plot(x, x**3, label="x**3") # Plot x vs x cubed, assigning a label for the legend
ax.legend() # Display the legend, using the labels defined in the plot calls
```

Notice how are legend overlaps some of the actual plot!

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **loc** are numerical codes for the various places the legend can be drawn. See the [documentation page](https://matplotlib.org/stable/) for details. Some of the most common **loc** values are:

```python
# Lots of options....

ax.legend(loc=1) # Place the legend in the upper right corner
ax.legend(loc=2) # Place the legend in the upper left corner
ax.legend(loc=3) # Place the legend in the lower left corner
ax.legend(loc=4) # Place the legend in the lower right corner

# .. many more options are available

# Most common to choose
ax.legend(loc=0) # Let Matplotlib decide the optimal location for the legend
fig # Display the figure, showing the effect of the last legend placement
```

## Setting colors, linewidths, linetypes

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes.

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

### Colors with MatLab like syntax

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
# MATLAB style line color and style
fig, ax = plt.subplots() # Create a figure and a single set of axes
ax.plot(x, x**2, 'b.-') # Plot x vs x squared with a blue line and dots
ax.plot(x, x**3, 'g--') # Plot x vs x cubed with a green dashed line
```

### Colors with the color= parameter

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
fig, ax = plt.subplots() # Create a figure and a single set of axes

ax.plot(x, x+1, color="blue", alpha=0.5) # Plot with a blue color and 50% opacity
ax.plot(x, x+2, color="#8B008B")        # Plot with a color specified by RGB hex code (DarkMagenta)
ax.plot(x, x+3, color="#FF8C00")        # Plot with a color specified by RGB hex code (DarkOrange)
```

### Line and marker styles

To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
fig, ax = plt.subplots(figsize=(12,6)) # Create a figure and a single set of axes with a specified figure size

ax.plot(x, x+1, color="red", linewidth=0.25) # Plot with red color and linewidth of 0.25
ax.plot(x, x+2, color="red", linewidth=0.50) # Plot with red color and linewidth of 0.50
ax.plot(x, x+3, color="red", linewidth=1.00) # Plot with red color and linewidth of 1.00
ax.plot(x, x+4, color="red", linewidth=2.00) # Plot with red color and linewidth of 2.00

# possible linestype options â-â, âââ, â-.â, â:â, âstepsâ
ax.plot(x, x+5, color="green", lw=3, linestyle='-') # Plot with green color, linewidth 3, and solid linestyle
ax.plot(x, x+6, color="green", lw=3, ls='-.') # Plot with green color, linewidth 3, and dash-dot linestyle
ax.plot(x, x+7, color="green", lw=3, ls=':') # Plot with green color, linewidth 3, and dotted linestyle

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50) # Plot with black color and linewidth 1.5, storing the Line2D object
line.set_dashes([5, 10, 15, 10]) # Set a custom dash pattern: 5 units line, 10 units space, 15 units line, 10 units space

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+') # Plot with blue color, solid line, and '+' markers
ax.plot(x+10, color="blue", lw=3, ls='--', marker='o') # Plot with blue color, dashed line, and 'o' (circle) markers
ax.plot(x+11, color="blue", lw=3, ls='-', marker='s') # Plot with blue color, solid line, and 's' (square) markers
ax.plot(x+12, color="blue", lw=3, ls='--', marker='1') # Plot with blue color, dashed line, and '1' (triangle_down) markers

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2) # Plot with purple, small circle markers
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4) # Plot with purple, medium circle markers
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red") # Plot with purple, large circle markers with red fill
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green"); # Plot with purple, large square markers with yellow fill, green edge, and thick edge
```

### Control over axis appearance

In this section we will look at controlling axis sizing properties in a matplotlib figure.

## Plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4)) # Create a figure and a 1x3 grid of subplots with a specified figure size

axes[0].plot(x, x**2, x, x**3) # Plot x vs x^2 and x vs x^3 on the first subplot
axes[0].set_title("default axes ranges") # Set the title for the first subplot

axes[1].plot(x, x**2, x, x**3) # Plot x vs x^2 and x vs x^3 on the second subplot
axes[1].axis('tight') # Automatically set tight limits for the second subplot axes
axes[1].set_title("tight axes") # Set the title for the second subplot

axes[2].plot(x, x**2, x, x**3) # Plot x vs x^2 and x vs x^3 on the third subplot
axes[2].set_ylim([0, 60]) # Set the Y-axis limits for the third subplot
axes[2].set_xlim([2, 5]) # Set the X-axis limits for the third subplot
axes[2].set_title("custom axes range"); # Set the title for the third subplot
```
